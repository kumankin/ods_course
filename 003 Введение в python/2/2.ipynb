```python
def operation_result(operator, a, b):
    a, b = int(a), int(b)
    if operator == '+':
        return(a + b)
    if operator == '-':
        return(a - b)
    if operator == '*':
        return(a * b)
    if operator == '//':
        return(a // b)
    if operator == '%':
        return(a % b)
    if operator == '**':
        return(a ** b)
```

```python
file_input = open('data/input_file.txt', 'r')
file_output = open('data/output_file.txt', 'w')

all_data_input = file_input.readlines()
```

```python
text_file_output = ''
for str_in_file_input in all_data_input:
    str_in_file_input_split = str_in_file_input.split('    ')
    text_file_output += str(operation_result(str_in_file_input_split[0], str_in_file_input_split[1], str_in_file_input_split[2].replace('/n', ''))) + ','
file_output.write(text_file_output[:-1])
file_output.close()
file_input.close()
```
