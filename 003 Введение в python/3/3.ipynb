```python
file_input_1 = open('data/import_file_2_1.txt', 'r')
all_data_input_1 = file_input_1.readlines()
file_input_1.close()

file_input_2 = open('data/import_file_2_2.txt', 'r')
all_data_input_2 = file_input_2.readlines()
file_input_2.close()

file_output = open('data/output_file.txt', 'w')
```

```python
text_file_output = ''
if len(all_data_input_1) == len(all_data_input_2):
    for i in range(len(all_data_input_1)):
        str_in_file_input_2_split = all_data_input_2[i].split(' ')
        a, b = int(str_in_file_input_2_split[0]), int(str_in_file_input_2_split[1].replace('/n', '')) + 1
        text_file_output += str(all_data_input_1[i].replace('/n', '')[a:b]) + ' '
file_output.write(text_file_output[:-1])
file_output.close()
```
