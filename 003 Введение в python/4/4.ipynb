```python
import json
file_periodic_table = open('data/periodic_table.json', 'r', encoding='utf-8')
periodic_table = json.load(file_periodic_table)
```

```python
file_input = open('data/import_file_3.txt', 'r')
all_data = file_input.read()
file_input.close()
```

```python
list_output = []
while len(all_data) > 0:
    for element in periodic_table:
        if all_data.endswith(element):
            all_data = all_data[:-len(element)]
            list_output.append(periodic_table[element])
            break
print(''.join(list_output[::-1]))
```
